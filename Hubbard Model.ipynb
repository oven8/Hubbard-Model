```julia
using ITensors
function two_component_BH(N::Int64;U=1.0,U12=-0.5,t1=0.5,t2=0.25,kwargs...)
    sites = siteinds("Boson",N;conserve_qns=true,conserve_number=false)
    op = OpSum()
    for i in 1:2:N-3
        op += -U/2, "N", i
        op += U/2, "N", i,"N",i
        op += -t1, "a", i, "a†", i + 2
        op += -t1, "a†", i, "a", i + 2
    end
    op += -U/2, "N", N-1
    op += U/2, "N", N-1,"N",N-1

    for i in 2:2:N-2
        op += -U/2, "N", i
        op += U/2, "N", i,"N",i
        op += -t2, "a", i, "a†", i + 2
        op += -t2, "a†", i, "a", i + 2
    end
    op += -U/2, "N", N
    op += U/2, "N", N,"N",N

    for i in 1:2:N-1
        op +=U12, "N", i,"N",i+1
    end
    MPO(op,sites;kwargs...)
end

two_component_BH(6)
```

```julia
using ITensors
function one_component_BH(M::Int64,sites;U=1.0,J=0.5,mu=0,V =0,kwargs...)
    sites = sites
    op = OpSum()
    for i in 1:M-1
        op += -mu -U/2, "N", i
        op += U/2, "N", i,"N",i
        op += -J, "a", i, "a†", i + 1
        op += -J, "a†", i, "a", i + 1
        op += V, "N", i,"N", i+1
    end
    op += -mu-U/2, "N", M
    op += U/2, "N", M,"N",M
    
    MPO(op,sites;kwargs...)
end

sites = siteinds("Boson",6;conserve_qns=true)
one_component_BH(6,sites)
```

```julia
using ITensors
function one_component_FH(M::Int64,sites;U=1.0,J=0.5,mu=0,kwargs...)
    sites = sites
    op = OpSum()
    for i in 1:M-1
        op += -mu, "n↑", i
        op += -mu, "n↓", i
        op += U, "n↑", i,"n↓",i
        op += -J, "c↑", i, "c†↑", i + 1
        op += -J, "c†↑", i, "c↑", i + 1
        op += -J, "c↓", i, "c†↓", i + 1
        op += -J, "c†↓", i, "c↓", i + 1
    end
    op += -mu, "n↑", M
    op += -mu, "n↓", M
    op += U, "n↑", M,"n↓",M
    
    MPO(op,sites;kwargs...)
end

sites = siteinds("Electron",6;conserve_qns=true)
one_component_FH(6,sites)
```
